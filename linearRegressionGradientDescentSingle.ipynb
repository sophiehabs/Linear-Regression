# Single Variable "home"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('home.csv',names=["size","bedroom","price"])
```

```python
#we need to normalize the features using mean normalization
my_data = (my_data - my_data.mean())/my_data.std()
```

```python
my_data.head()
```

```python
#setting the matrixes
X = my_data.iloc[:,0:1] # use first column of data
ones = np.ones([X.shape[0],1])
X = np.concatenate((ones,X),axis=1)
```

```python
y = my_data.iloc[:,2:3].values # third column as target variable
theta = np.zeros([1,2]) # change the shape of theta to match the number of features
```

```python
#computecost
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))

def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost
```

```python
#set hyper parameters
#alpha = 0.01
#iters = 1000
alpha = [0.01, 0.03, 0.1, 0.3, 1]
iters = [20, 30, 50, 100, 200, 500, 1000, 5000, 10000]

for alph in alpha:
    for it in iters:
        g, cost = gradientDescent(X, y, theta, it, alph)
        print("For alpha = ", alph, "and iterations = ", it, "g = ", g)

        plt.scatter(my_data.iloc[:,0:1], y, c='b', marker='+')
        plt.xlabel("Size")
        plt.ylabel("Price")

        # make predictions using the optimized theta
        predictions = X @ g.T

        # plot the prediction
        plt.plot(my_data.iloc[:,0:1], predictions, c='r')
        plt.show()
```

Cost reduces as the number of iterations increases. The cost has reached a minimum value after 1000 iterations and does not change significantly for more iterations. However, increasing the number of iterations also increases the computation time.

```python
for alph in alpha:
    for it in iters:
        g, cost = gradientDescent(X, y, theta, it, alph)
        print("For alpha = ", alph, "and iterations = ", it, "g = ", g)
```
