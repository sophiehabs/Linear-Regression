# Single Variable "dataWO"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('dataWO.csv' ,names=["size", "price"])
```

```python
my_data.head()
```

```python
#we need to normalize the features using mean normalization
my_data = (my_data - my_data.mean())/my_data.std()
```

```python
my_data.head()
```

```python
# Extract both columns as NumPy arrays
X = my_data.iloc[:, 0].values
y = my_data.iloc[:, 1].values

# Reshape X and y
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

# concatenate
X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)

# Print shapes of X and y
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# theta to zeros
theta = np.zeros((1, 2))

# Print shape of theta
print("theta shape:", theta.shape)
```

```python
# Define computeCost function
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))

# Define gradientDescent function
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    return theta, cost
```

```python
# Set hyper parameters
alpha = [0.01, 0.03, 0.1, 0.3, 1]
iters = [20, 30, 50, 100, 200, 500, 1000, 5000, 10000]

for alph in alpha:
    for it in iters:
        g, cost = gradientDescent(X, y, theta, it, alph)
        print("For alpha = ", alph, "and iterations = ", it, "g = ", g)

        plt.scatter(my_data.iloc[:,0:1], y, c='b', marker='+')
        plt.xlabel("Size")
        plt.ylabel("Price")

        # make predictions using the optimized theta
        predictions = X @ g.T

        # plot the prediction
        plt.plot(my_data.iloc[:,0:1], predictions, c='r')
        plt.show()
```
